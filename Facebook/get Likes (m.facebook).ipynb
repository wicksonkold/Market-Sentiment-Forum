```python
df_reaction = pd.DataFrame(columns=["Total Reaction",
                                    "Like",
                                    "Angry",
                                    "Heart",
                                    "HaHa",
                                    "Add Oil", 
                                    "Cry", 
                                    "WOW"])
def sort_reactions(result_list):
    #defual
    total = 0
    like = 0
    angry = 0
    heart = 0
    haha = 0
    add_oil = 0
    cry = 0
    wow = 0
    
    for temp in result_list:
        temp_result = temp.split(' ')   
        if temp_result[1] == "人對這則貼文傳達了心情":
            total = int(temp_result[0])
        if temp_result[1] == "人表示讚":
            like =  int(temp_result[0])
        if temp_result[1] == "人表示怒":
            angry = int(temp_result[0])
        if temp_result[1] == "人表示大心":
            heart =  int(temp_result[0])
        if temp_result[1] == "人表示哈":
            haha =  int(temp_result[0])
        if temp_result[1] == "人表示加油":
            add_oil = int(temp_result[0])
        if temp_result[1] == "人表示嗚":
            cry = int(temp_result[0])
        if temp_result[1] == "人用哇傳達了心情":
            wow = int(temp_result[0])
            
    df_reaction.loc[len(df_reaction)] = [total, like, angry, heart, haha, add_oil, cry, wow]
    return 
```

```python
for link in reaction_like:
    driver.get(link)
    html = driver.page_source
    soup = BeautifulSoup(html)
    result_list = []
    for i in soup.find_all("span",{"class":"_10tn"}):
        temp_result = i.find('span')['aria-label']
        print(temp_result)
        result_list.append(temp_result)
        print("="*20)
    sort_reactions(result_list)
```
